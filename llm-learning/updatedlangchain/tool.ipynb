An LLM can call a tools to fetch API or things from the web. Tools are generally important


1. Schema, Including  the name a description, and/or argument definition(often JSON schema)
2. A function or coroutine to execute.

```python
import os
from dotenv import load_dotenv

load_dotenv()

from langchain.chat_models import init_chat_model

api_key = os.get("GROQ_API_KEY")

if api_key is None:
    raise ValueError("API is not available for this model")

model = init_chat_model("qwen/qwen3-32b")

response = model.invoke("what is a noun? ")

response 
```

```python

from langchain.tools import tool

@tools
def get_weather(location:str)-> str:
    """Get weather at a location"""
    return f"Its sunny in this {location}"

model_with_tools = model.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke("whats the weather in Boston ?")
print(response)
```
