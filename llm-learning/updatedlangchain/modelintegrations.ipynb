### Model integration With Google Gemini Groq Open AI

#### Calling OpenAI modal -- using init_chat_models


```python
import os

from dotenv import load_dotenv
load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
if api_key is None:
    raise ValueError("OPENAI_API_KEY not found in environment variables")


from langchain.chat_models import init_chat_model

model = init_chat_model("gpt-4.1")

response = model.invoke("Why do parrots talk?")
response.content
```

#### Calling google gemini modal -- using init_chat_modals

```python
import os

from dotenv import load_dotenv
load_dotenv()

api_key: str | None = os.getenv("GOOGLE_API_KEY")
if api_key is None:
    raise ValueError("GOOGLE_API_KEY not found in environment variables")


from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-3-flash-preview")

response = model.invoke("Why do parrots talk?")

response.content[0]['text']
```

#### Another way to use Groq is using the ChatGroq

```python
from langchain_groq import ChatGroq

model = ChatGroq(model="qwen/qwen3-32b")

response = model.invoke("Why do parrots talk?")
response.content
```

### Streaming

*Streaming* (Real-Time Generation): Optimal UX
When you need immediate user feedback, choose streaming. The model returns its response token-by-token as it is calculated, minimizing perceived latency. This keeps users engaged by delivering a smooth, live-typing experience.


```python
from langchain_groq import ChatGroq

model = ChatGroq(model="qwen/qwen3-32b")

response = model.stream("Why do parrots talk?")

for chunk in response:
    print(chunk.content, end="", flush=True)
```

### Batch 

*Batching (Bulk Evaluation)*: Optimal Budget
When you need throughput efficiency or are managing offline pipelines, use batching. You group multiple distinct prompts together in a single request, which cloud providers often offer at a significant discount (sometimes 50% cheaper) because they can process the workload asynchronously.

```python
from langchain_groq import ChatGroq

model = ChatGroq(model="qwen/qwen3-32b")

response = model.batch(
    inputs = [
        "what is the fasted  growing economy in africa",
        "what is the reason for the fuel hike in 2026",
        "what is the current state in south africa Xenxophobia"
    ],
    config = {
         "max_concurrency": 5
    }
)

for chunk in response:
    print(chunk.content)
```
